```python
import pandas as pd

df = pd.read_csv("../data/access_logs.csv", parse_dates=["timestamp"])
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
df["hour"] = df["timestamp"].dt.hour
df["day_of_week"] = df["timestamp"].dt.day_name()

df[["timestamp", "hour", "day_of_week"]].head()
```

```python
df.to_csv("../data/access_logs_cleaned.csv", index=False)
print("Cleaned dataset saved.")
```
