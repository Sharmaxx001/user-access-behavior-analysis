```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/access_logs_cleaned.csv", parse_dates=["timestamp"])
df.head()
```

```python
df.shape
```

```python
top_users = df["user_id"].value_counts().head(10)

top_users.plot(kind="bar", title="Top 10 Users by Access Count")
plt.xlabel("User ID")
plt.ylabel("Access Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(x="hour", data=df)
plt.title("Access Activity by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Access Events")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df["failed_logins"], bins=10)
plt.title("Failed Login Distribution")
plt.xlabel("Failed Login Count")
plt.tight_layout()
plt.show()
```

```python
df["location"].value_counts().plot(kind="bar", title="Access Events by Location")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
